# 파생변수 생성
1. 원가, 단가, 판매단위 값을 기준으로
2. 총수익, 총비용, 총이윤 이라는 새로운 columns를 생성
3. 총수익 = 단가 *판매단위
4. 총비용 = 원가 *판매단위
5. 총이윤 = 총수익 - 총비용

```python
# 새로운 컬럼을 추가하는 방법
# 데이터프레임명[새로운 컬럼의 이름] = 컬럼에 들어갈 값 (Series, list)
# 단가 컬럼 * 판매단위 컬럼
sales["총수익"] = sales['Unit Price'] * sales['Units Sold']
```

```python
sales['Unit Price'] #데이터 형태가 Series (index값이 존재하는 게 Series)
```

```python
sales.head()
# 뒤에 총수익 컬럼이 생김
```

```python
# 총비용, 총이윤 만들기
sales["총비용"] = sales["Unit Cost"] * sales['Units Sold']
sales["총이윤"] = sales['총수익'] - sales['총비용']
sales.head()
```
