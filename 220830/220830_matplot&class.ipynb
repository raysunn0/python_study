변수랑 식을 합쳐놓은 하나의 영역  
변수 : 반찬통, 그 통에 클래스라는걸 그냥 집어 넣은 것.  
self는 반찬통의 주소, 반찬통이 어디 있는지.

```python
class Class_3():
    def __init__(self, input_x, input_y):
        self.x = input_x #10
        self.y = input_y #4
    def add(self): #class 안에 add 함수를 넣은 것. 지금 매개 변수 안만들어져 있는 것.
        self.result = self.x + self.y
        return self.result
    def subtract(self):
        self.result = self.x - self.y
        return self.result
```

```python
c = Class_3(10, 4) #c라는 변수 명에 class_3을 넣고 10과 4를 호출할 것.
```

c라는 반찬통에 집어넣은 것임.  
변수를 꺼내서 확인해보거나 함수를 꺼내서 함수의 값을 확인해보거나.

변수 c, c라는 공간에 뭐를 쓰겠다. c라는 이름에 class_3를 넣어둔 것.  

c1, c2, c3, c4 모두 class_3를 넣어뒀지만 class는 이 안에 변수를 한 번 더 지정할 수 있음.  
c1일 때는 1, 2 / c2일 때는 3, 4 이렇게 한 번 더 넣는 것. 이런 형태로 변수 따로 지정할 수 있음.  
변수 따로 지정이 되면 안에 들어가는 함수도 반환값이 달라짐.    
=> 변수와 함수 모두 들어가 있는 게 class  

```python
print(c.add()) #class 안에 넣어뒀기 때문에 print하면 더한 값이 나옴.
print(c.subtract())
```

1. class 생성
2. class 생성하는 과정에서 인자 값을 2개 (a, b) 받고
3. 새로운 메서드(함수) 하나 생성해서
4. 생성할 때 받은 인자 값 a, b
5. a와 b 사이의 값들의 합계를 출력하는 함수 생성

```python
class Class_10():
    def __init__(self, a, b):
        sum = 0
        for i in range(a, b+1):
            if a < b :
                sum += i
            else :
                for i in range(b, a+1):
                    sum += i
        print(sum)
```

```python
test = Class_10(4, 1)
```

해답

```python
class Class_10() : #상속이 없으면 ()생략해도 됨
    def __init__(self, a, b):
        self.a = a
        self.b = b

    def func_sum(self):
        self.sum = 0
        for i in range(min(self.a, self.b)+1, max(self.a, self.b), 1): #둘 중 작은 값
            self.sum += i
        return self.sum
```

```python
a = Class_10(5, 1)
```

```python
print(a.func_sum())
```
