1. continue
- 반복문으로 돌아간다.
- continue에서는 밑에 남아있는 코드를 실행시키지 않고 반복문으로 돌아간다.

- 반복문을 종료한다.

```python
for i in range(1, 100, 1):
    if i < 10:
        continue
    elif i > 15:
        break

    print(i, end=', ') # 두 조건이 거짓이면 실행
```

```python
# 반복문을 이용해서 1부터 100까지 반복
# 25부터 45까지의 합계를 출력

# case 1, continue break 연습
sum = 0

for i in range(1, 100):
    if i < 25:
        continue
    elif i > 45:
        break
    sum += i
print(sum)
```

```python
# case 2, while 문
i = 1
sum = 0

while i < 100:
    if i < 25:
        i += 1
        continue
    elif i > 45:
        break
    sum += i
    i += 1
print(sum)
```

```python
# case 3, 원래 쓰는 방법
sum = 0
for i in range(1, 100, 1):
    if i >= 25 and i <= 45:
        sum += i
    elif i > 45:
        break
print(sum)
```

# 함수 
1. 변수는 데이터의 값을 가진 데이터의 공간
2. 함수는 데이터의 값이 아니라 입력받은 데이터를 가지고 뭔가 행동을 하여 결과값을 돌려주는 코드
3. def 함수명 (매개변수)
4. 함수를 호출할 때는 함수명( 인자 )
5. 일반적으로는 매개변수의 개수와 인자의 개수가 동일해야 한다.

```python
# 매개변수가 존재하지 않는 함수
def func_1():
    return "hello world"

print(func_1())
```

```python
# 매개변수가 존재하는 함수
#result = ""

def func_2(x, y):
    if type(x) == int and type(y) == int:
        result = x ** y
    else :
        result = "입력받은 값의 타입이 숫자형이 아닙니다."
    return result

print(func_2(5, 4))
print(func_2("a", "b"))
```

```python
# 매개변수에 기본값이 존재하는 경우
def func_3(x, y = 4):
    result = x+y
    return result

print(func_3(3))
```

```python
def func_4(x, y, z):
    result = (x + y) * z
    return result

print(func_4(5, 2, 3))
```

# Class
1. 변수, 함수의 집합체
2. Class 보통 이름을 대문자로 시작해서 생성
3. self라는 매개변수 -> 자기 자신을 뜻 -> Class가 저장되어 있는 공간  
#self 변수를 통해 어느 위치의 class인지 ㅇㅇ
4. '__init__' 함수 -> Class가 생성이 될 때, 단 한 번만 실행이 되는 함수
5. '__init__' 함수는 필요 시에만 작성

```python
# Class 선언
class Class_1():
    def __init__(self):
        self.result = 'Class가 생성되었습니다.'

    def output(self):
        return self.result
```

```python
# Class 생성
Class_a = Class_1()
print(Class_a.output())
```

```python
# 변수와 함수가 존재하는 클래스 선언
class Class_2():
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def output(self):
        self.result = self.x + self.y
        return self.result
```

```python
# class_2 생성
class_b = Class_2(5, 4)
class_c = Class_2(10, 3)

print(class_b.output())
print(class_c.output())
```

### 공백제거, 대문자 변경을 해주는 Class 생성
1. init함수에는 파일 경로를 매개변수 이용해 값을 받아오는 작업 진행
2. 함수 생성 -> 공백제거, 대문자로 변환 리턴 -> 이 함수에서 필요한 매개변수 : 특정 컬럼 받아오기
3. 데이터프레임을 리턴

```python
# 공백제거 하는 함수를 Class에 담아보기
class Class_4():
    def __init__ (self, _path): #_path : _는 입력받은 값을 의미
        self.df = pd.read_csv(_path)

    def change(self, _column):
        # 공백제거
        self.df[_column] = self.df[_column].str.replace(" ", "") # .str.upper()여기 붙여도 됨
        # 대문자로 변환
        self.df[_column] = self.df[_column].str.upper()
        return self.df

    def review(self):
        return self.df 
```

```python
class_df = Class_4("../csv/uriage.csv")
class_df.review()
```

```python
class_df.change('item_name') #()에 column 이름 넣으면 실행됨
```

```python
class Class_df():
    def __init__ (self, _path): #_path : _는 입력받은 값을 의미
        self.df = pd.read_csv(_path)

    def change(self, _column):
        # 공백제거
        self.df[_column] = self.df[_column].str.replace(" ", "") # .str.upper()여기 붙여도 됨
        # 대문자로 변환
        self.df[_column] = self.df[_column].str.upper()
        return self.df

    def change_dt(self, _column):
        self.df[_column] = pd.to_datetime(self.df[_column])
        return self.df

    def change_ymd(self, _column):
        self.df['year'] = self.df[_column].dt.strftime("%Y")
        self.df['month'] = self.df[_column].dt.strftime("%m")
        self.df['day'] = self.df[_column].dt.strftime("%d")
        return self.df

    def review(self):
        return self.df 
```
