```python
class Class_10() : #상속이 없으면 ()생략해도 됨
    def __init__(self, a, b):
        self.a = a
        self.b = b

    def func_sum(self):
        self.sum = 0
        for i in range(min(self.a, self.b)+1, max(self.a, self.b), 1): #둘 중 작은 값
            self.sum += i
        return self.sum


a = Class_10(5, 1)

print(a.func_sum()) #9
```

### datetime format
- %y : 년도를 축약해서 표시 ('22')
- %Y : 년도의 전체를 표시 ('2022')
- %b : 축약이 된 월의 이름 ('Mar')
- %B : 월의 이름 ('March')
- %m : 월을 숫자로 표시 ('08')
- %d : 일을 숫자로 표시 ('31')
- %H : 시를 24시를 기준으로 표시
- %I : 시를 12시를 기준으로 표시 
- %M : 분을 표시 => 소문자 m이 월에 있기 때문
- %s : 초를 표시
- %a : 요일의 이름을 축약해서 표시
- %A : 요일의 이름을 표시
- %w : 요일을 숫자로 표시 (일 : 0, 월 : 1, ... 토 : 6)
